```python
import tensorflow as tf

mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
resource_path = "./resource/abccodes/"
```

```python
with open(resource_path + "x_train.txt", 'r') as file:
    x_train_files = file.readlines()
```

```python
x_train = []
for i in x_train_files:
    temp = i.split(',')
    temp_list = []
    for j in temp:
        temp_list.append(int(j))
    x_train.append(temp_list)
```

```python
with open(resource_path + "y_train.txt", 'r') as file:
    y_train_files = file.readlines()
```

```python
y_train = []
for i in y_train_files:
    y_train.append(int(i))
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(12,)),
    tf.keras.layers.Dense(144, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(144, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(144, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(144, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(144, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(144, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(144, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(144, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(36, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
x_train_set = x_train[0:18000]
```

```python
x_test_set = x_train[18000:]
```

```python
y_train_set = y_train[0:18000]
```

```python
y_test_set = y_train[18000:]
```

```python
model.fit(x_train_set, y_train_set, epochs=5)
```

```python
model.evaluate(x_test_set, y_test_set)
```

```python
p_array = model.predict(x_test)
```

```python
p_array = list(p_array)
```

```python
for i in p_array:
    i = list(i)
    for j in range(len(i)):
        if(max(i) == i[j]):
            print(j)
```

```python
model.evaluate(x_test, y_test)
```
