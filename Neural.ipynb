```python
resource_path = "./resource/abccodes/"
```

```python
with open(resource_path + "Total.txt", 'r') as file:
    codes = file.readlines()
```

```python
# 원본
codes
```

```python
type(codes)
```

```python
codes_x1 = []
```

```python
for i in codes:
    if('"' in i or 'X:' in i or 'K:' in i):
        codes_x1.append(i.split('|'))
```

```python
# 색인, 음계, 코드부분만 때오고 마디별로 분할
codes_x1
```

```python
codes_x2 = []
```

```python
for i in codes_x1:
    for j in i:
        j = j.replace(' "', ' *"')
        j = j.replace(' -"', ' *"')
        codes_x2.append(j)
```

```python
# 한마디안에 코드 2개 들어있는부분 전처리
codes_x2
```

```python
codes_x3 = []
```

```python
for i in codes_x2:
    codes_x3.append((i.split(' *')))
```

```python
# 마디 2개있는 부분 분할
codes_x3
```

```python
codes_x4 = []
```

```python
for i in codes_x3:
    for j in i:
        if('"' in j or j[0:2] == 'X:' or j[0:2] == 'K:'):
            codes_x4.append(j)
```

```python
# 분할된 리스트에서 다시 색인과 코드, 음계가 포함된 원소만 들고옴
codes_x4
```
